```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("../lecture_07/panda.jpg")
```

```python
img = cv2.resize(img, (img.shape[1]//10, img.shape[0]//10))
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
def apply_filter(gray, fil):
    res = np.zeros([gray.shape[0]-(fil.shape[0]-1), gray.shape[1]-(fil.shape[1]-1)])
    for r in range(res.shape[0]):
        for c in range(res.shape[1]):
            cut = gray[r:r+fil.shape[0], c:c+fil.shape[1]]
            res[r][c] = np.mean(cut * fil)
            
    return res
```

```python
out = apply_filter(gray, np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]]))
```

```python
plt.imshow(out, cmap="gray")
```

```python
plt.imshow(gray, cmap="gray")
```

```python
def apply_pool(gray, shape):
    res = np.zeros([gray.shape[0] // shape[0], gray.shape[1] // shape[1]])
    for r in range(0, res.shape[0]):
         for c in range(0, res.shape[1]):
                r_start = r * shape[0]
                c_start = c * shape[1]
                cut = gray[r_start:r_start+shape[0], c_start:c_start+shape[1]]
                res[r][c] = np.max(cut)
    return res
```

```python
out = apply_pool(gray, (2, 2))
```

```python
plt.imshow(out, cmap="gray")
```
