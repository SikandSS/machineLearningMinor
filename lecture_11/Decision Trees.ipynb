```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = []

for i in range(1, 50):
    red = i
    black = 50 - i
    
    entropy = - ((red/50)*np.log2(red/50) + (black/50)*np.log2(black/50))
    y.append(entropy)
    
plt.plot(y)
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=100)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X_left = X[np.logical_or((y == 0), (y == 1))]
```

```python
plt.scatter(X_left[:, 0], X_left[:, 1])
```
