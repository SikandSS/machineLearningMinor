```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
si = np.sin(x)
co = np.cos(x)
```

```python
plt.figure()
plt.plot(x, si, label="sin")
plt.plot(x, co, label="cos")
plt.xlabel("angles")
plt.ylabel("values")
plt.title("A graph between sin and cos")
plt.xlim(0, 2*np.pi)
plt.ylim(-2, 2)
# plt.legend()
plt.show
```
